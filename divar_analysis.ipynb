```python
# ============================================
# EXTRACT INFO FROM FARSI TEXT
# ============================================
import re

def persian_to_english_num(text):
    """Convert Persian/Arabic numerals to English"""
    if pd.isna(text):
        return text
    persian_nums = '۰۱۲۳۴۵۶۷۸۹'
    arabic_nums = '٠١٢٣٤٥٦٧٨٩'
    english_nums = '0123456789'
    text = str(text)
    for i, p in enumerate(persian_nums):
        text = text.replace(p, english_nums[i])
    for i, a in enumerate(arabic_nums):
        text = text.replace(a, english_nums[i])
    return text

def extract_rooms_from_text(text):
    """Extract room count from Farsi text"""
    if pd.isna(text):
        return None
    text = persian_to_english_num(str(text))
    
    # Patterns for room count: "2خواب", "۲ خوابه", "3خ", "سه خواب"
    patterns = [
        r'(\d+)\s*خواب',        # 2خواب, 2 خواب
        r'(\d+)\s*خ\b',          # 2خ
        r'(\d+)\s*اتاق',         # 2اتاق
    ]
    
    for pattern in patterns:
        match = re.search(pattern, text)
        if match:
            return int(match.group(1))
    
    # Word numbers
    word_nums = {'یک': 1, 'دو': 2, 'سه': 3, 'چهار': 4, 'پنج': 5}
    for word, num in word_nums.items():
        if re.search(f'{word}\s*خواب', text):
            return num
    
    return None

def extract_floor_from_text(text):
    """Extract floor number from Farsi text"""
    if pd.isna(text):
        return None
    text = persian_to_english_num(str(text))
    
    # Patterns: "طبقه3", "طبقه ۳"
    patterns = [
        r'طبقه\s*(\d+)',
        r'(\d+)\s*طبقه',
    ]
    
    for pattern in patterns:
        match = re.search(pattern, text)
        if match:
            return int(match.group(1))
    return None

def extract_size_from_text(text):
    """Extract building size from Farsi text"""
    if pd.isna(text):
        return None
    text = persian_to_english_num(str(text))
    
    # Patterns: "60 متر", "۶۰متر", "100متری"
    patterns = [
        r'(\d+)\s*متر',
    ]
    
    for pattern in patterns:
        match = re.search(pattern, text)
        if match:
            size = int(match.group(1))
            # Sanity check: size should be between 10-2000 sqm
            if 10 <= size <= 2000:
                return size
    return None

# Test extraction on sample titles
print("Testing extraction on sample titles:")
print("=" * 60)
for title in df_clean['title'].dropna().head(15):
    rooms = extract_rooms_from_text(title)
    floor = extract_floor_from_text(title)
    size = extract_size_from_text(title)
    print(f"Title: {title}")
    print(f"  → Rooms: {rooms}, Floor: {floor}, Size: {size}")
    print()
```

```python
# ============================================
# IMPROVED EXTRACTION FUNCTIONS
# ============================================

def extract_rooms_from_text_v2(text):
    """Extract room count from Farsi text - improved"""
    if pd.isna(text):
        return None
    text = persian_to_english_num(str(text))
    
    # Direct word matches first (more reliable)
    word_patterns = {
        'تک\s*خواب': 1,
        'یک\s*خواب': 1,
        'دو\s*خواب': 2,
        'سه\s*خواب': 3,
        'چهار\s*خواب': 4,
        'پنج\s*خواب': 5,
    }
    
    for pattern, num in word_patterns.items():
        if re.search(pattern, text):
            return num
    
    # Patterns for room count: "2خواب", "۲ خوابه", "3خ"
    patterns = [
        r'(\d+)\s*خواب',        # 2خواب, 2 خواب
        r'(\d+)\s*خ(?:\s|$|/)',  # 2خ (followed by space, end, or /)
    ]
    
    for pattern in patterns:
        match = re.search(pattern, text)
        if match:
            num = int(match.group(1))
            if 0 <= num <= 10:  # Sanity check
                return num
    
    return None

# Test improved function
print("Testing improved room extraction:")
test_titles = [
    'منظریه ۱۵۰متر ۳خ',
    '۷۰متر دوخواب /فول بازسازی',
    '۶۵ متر / تک خواب / پارکینگ',
    'فروش آپارتمان 143 متر 3 خواب',
    'آپارتمان ۸۰ متری ۲ خوابه',
]
for t in test_titles:
    print(f"  {t} → {extract_rooms_from_text_v2(t)} rooms")
```

```python
# ============================================
# IMPLEMENT SMART CONTEXT-AWARE IMPUTATION
# ============================================

print("=" * 60)
print("🧠 SMART IMPUTATION RULES BASED ON DATA ANALYSIS")
print("=" * 60)

# Reload and prep fresh data
df_smart_v2 = pd.read_csv('/Users/yashakhorsandi/Desktop/پروژه دیوار/real_estate_ads.csv', low_memory=False)
df_smart_v2 = df_smart_v2[(df_smart_v2['cat2_slug'] == 'residential-sell') & (df_smart_v2['city_slug'] == 'tehran')]
df_smart_v2 = df_smart_v2[df_smart_v2['cat3_slug'].isin(['apartment-sell', 'house-villa-sell'])]

# Drop irrelevant columns
drop_cols = ['rent_mode', 'rent_value', 'rent_to_single', 'rent_type', 
             'transformable_rent', 'transformed_rent', 'rent_price_on_regular_days', 
             'rent_price_on_special_days', 'rent_price_at_weekends',
             'rent_credit_transform', 'credit_value', 'credit_mode',
             'extra_person_capacity', 'regular_person_capacity', 'cost_per_extra_person',
             'transformed_credit', 'transformable_price', 'transformable_credit',
             'property_type', 'is_transformable', 'transformable_size', 'transformable_fee',
             'rent_base', 'rent_prepaid', 'rent_prepaid_percentage',
             'is_convertible', 'is_negotiable', 'has_swap', 'rent_daily_price', 'rent_discount',
             'credit', 'capacity', 'capacity_mode', 'capacity_value']
df_smart_v2 = df_smart_v2.drop(columns=[c for c in drop_cols if c in df_smart_v2.columns])

# Drop >90% missing columns  
missing_pct = df_smart_v2.isnull().mean() * 100
cols_high_missing = missing_pct[missing_pct > 90].index.tolist()
df_smart_v2 = df_smart_v2.drop(columns=cols_high_missing)

# Drop text columns
df_smart_v2 = df_smart_v2.drop(columns=['description', 'title'], errors='ignore')

# Convert to numeric where needed
df_smart_v2['total_floors_num'] = pd.to_numeric(df_smart_v2['total_floors_count'], errors='coerce')
df_smart_v2['floor_num'] = pd.to_numeric(df_smart_v2['floor'], errors='coerce')

print(f"Starting with {len(df_smart_v2)} rows")

# ==========================================
# RULE 1: ELEVATOR based on total floors
# ==========================================
print("\n📌 Rule 1: Elevator based on total floors")
print("   - 5+ floors: likely has elevator (98%+)")
print("   - 4 floors: coin flip (45%)")
print("   - 1-3 floors: likely no elevator (~25%)")

def impute_elevator(row):
    if pd.notna(row['has_elevator']):
        return row['has_elevator']
    
    floors = row['total_floors_num']
    if pd.isna(floors):
        return None  # Can't determine
    
    if floors >= 5:
        return True  # 98%+ have elevator
    elif floors <= 3:
        return False  # ~25% have elevator
    else:  # 4 floors - 45% have elevator
        return None  # Leave for mode imputation

before_elev = df_smart_v2['has_elevator'].isna().sum()
df_smart_v2['has_elevator'] = df_smart_v2.apply(impute_elevator, axis=1)
after_elev = df_smart_v2['has_elevator'].isna().sum()
print(f"   Imputed {before_elev - after_elev} elevator values")

# ==========================================
# RULE 2: WAREHOUSE based on building size
# ==========================================
print("\n📌 Rule 2: Warehouse based on building size")
print("   - >120 sqm: likely has warehouse (96%+)")
print("   - 80-120 sqm: likely has warehouse (93%)")
print("   - <50 sqm: lower probability (81%)")

def impute_warehouse(row):
    if pd.notna(row['has_warehouse']):
        return row['has_warehouse']
    
    size = row['building_size']
    if pd.isna(size):
        return None
    
    if size >= 120:
        return True  # 96%+ have warehouse
    elif size >= 80:
        return True  # 93% have warehouse
    elif size >= 50:
        return True  # 87% have warehouse
    else:
        return True  # Even small ones have 81%

before_wh = df_smart_v2['has_warehouse'].isna().sum()
df_smart_v2['has_warehouse'] = df_smart_v2.apply(impute_warehouse, axis=1)
after_wh = df_smart_v2['has_warehouse'].isna().sum()
print(f"   Imputed {before_wh - after_wh} warehouse values")

# ==========================================
# RULE 3: PARKING based on property type
# ==========================================
print("\n📌 Rule 3: Parking (~80% overall have parking)")

def impute_parking(row):
    if pd.notna(row['has_parking']):
        return row['has_parking']
    return True  # 80% have parking, default to True

before_park = df_smart_v2['has_parking'].isna().sum()
df_smart_v2['has_parking'] = df_smart_v2.apply(impute_parking, axis=1)
after_park = df_smart_v2['has_parking'].isna().sum()
print(f"   Imputed {before_park - after_park} parking values")

# ==========================================
# RULE 4: BALCONY based on building size
# ==========================================
print("\n📌 Rule 4: Balcony based on building size")
print("   - >80 sqm: very likely (94%+)")
print("   - 50-80 sqm: likely (84%)")
print("   - <50 sqm: less likely (67%)")

def impute_balcony(row):
    if pd.notna(row['has_balcony']) and row['has_balcony'] != 'unselect':
        return row['has_balcony']
    
    size = row['building_size']
    if pd.isna(size):
        return None
    
    if size >= 80:
        return 'true'  # 94%+ have balcony
    elif size >= 50:
        return 'true'  # 84% have balcony  
    else:
        return 'false'  # 67% - closer to uncertain

before_bal = (df_smart_v2['has_balcony'].isna() | (df_smart_v2['has_balcony'] == 'unselect')).sum()
df_smart_v2['has_balcony'] = df_smart_v2.apply(impute_balcony, axis=1)
after_bal = df_smart_v2['has_balcony'].isna().sum()
print(f"   Imputed {before_bal - after_bal} balcony values")

print("\n✅ Smart context-aware imputation complete!")
```

```python
# ============================================
# Define NLP Feature Extraction Patterns
# ============================================

def extract_nlp_features(text):
    """Extract multiple features from Persian text"""
    if pd.isna(text) or text == '':
        return pd.Series({
            'is_north_facing': False,
            'is_south_facing': False,
            'is_east_facing': False,
            'is_west_facing': False,
            'is_two_sided': False,
            'is_well_lit': False,
            'has_view': False,
            'is_newly_built': False,
            'is_renovated': False,
            'is_luxury': False,
            'is_fully_equipped': False,
            'is_never_used': False,
            'is_dead_end': False,
            'has_good_layout': False,
            'has_single_deed': False,
            'has_roof_garden': False,
            'has_lobby': False,
            'has_caretaker': False,
            'has_video_intercom': False,
            'has_built_in_closet': False,
            'has_remote_door': False
        })
    
    text = str(text).lower()
    
    return pd.Series({
        # Orientation / Direction
        'is_north_facing': bool(re.search(r'شمالی|رو به شمال|نما شمال', text)),
        'is_south_facing': bool(re.search(r'جنوبی|رو به جنوب|نما جنوب', text)),
        'is_east_facing': bool(re.search(r'شرقی|رو به شرق', text)),
        'is_west_facing': bool(re.search(r'غربی|رو به غرب', text)),
        'is_two_sided': bool(re.search(r'دوبر|دو بر|دو نبش|دونبش|سه بر|سه‌بر', text)),
        
        # Light & View
        'is_well_lit': bool(re.search(r'نورگیر|نور عالی|فول نور|پر نور|نور خوب|تمام نور', text)),
        'has_view': bool(re.search(r'ویو|چشم.?انداز|رو به پارک|منظره|دید', text)),
        
        # Condition & Quality
        'is_newly_built': bool(re.search(r'نوساز|نو ساز|صفر|تازه ساخت', text)),
        'is_renovated': bool(re.search(r'بازسازی|باز سازی|نوسازی شده|رنوویت', text)),
        'is_luxury': bool(re.search(r'لوکس|لاکچری|مجلل|شیک|اعیانی', text)),
        'is_fully_equipped': bool(re.search(r'فول امکانات|فول|تمام امکانات|کامل', text) and not re.search(r'فول نور', text)),
        'is_never_used': bool(re.search(r'کلید نخورده|کلیدنخورده|صفر کیلومتر', text)),
        
        # Location Features
        'is_dead_end': bool(re.search(r'بن.?بست|کوچه بسته|انتهای کوچه', text)),
        'has_good_layout': bool(re.search(r'خوش نقشه|نقشه عالی|نقشه خوب', text)),
        
        # Legal Status
        'has_single_deed': bool(re.search(r'تک برگ|تکبرگ|سند تک', text)),
        
        # Additional Amenities (not in structured data)
        'has_roof_garden': bool(re.search(r'روف|روف گاردن|بام سبز|پشت بام', text)),
        'has_lobby': bool(re.search(r'لابی|لابی مجلل|لابی مان', text)),
        'has_caretaker': bool(re.search(r'سرایدار|نگهبان|نگهبانی', text)),
        'has_video_intercom': bool(re.search(r'آیفون تصویری|ایفون تصویری|آیفن تصویری', text)),
        'has_built_in_closet': bool(re.search(r'کمد دیواری|کمددیواری|کمد.?دار', text)),
        'has_remote_door': bool(re.search(r'ریموت|درب ریموت|پارکینگ ریموت', text))
    })

print("NLP feature extraction function defined!")
print("\nFeatures to extract:")
features = ['is_north_facing', 'is_south_facing', 'is_east_facing', 'is_west_facing', 
            'is_two_sided', 'is_well_lit', 'has_view', 'is_newly_built', 'is_renovated',
            'is_luxury', 'is_fully_equipped', 'is_never_used', 'is_dead_end', 
            'has_good_layout', 'has_single_deed', 'has_roof_garden', 'has_lobby',
            'has_caretaker', 'has_video_intercom', 'has_built_in_closet', 'has_remote_door']
for f in features:
    print(f"  - {f}")
```

```python
# ============================================
# Reload original data to get title/description for all rows
# ============================================

# Reload original CSV
df_original_reload = pd.read_csv('/Users/yashakhorsandi/Desktop/پروژه دیوار/real_estate_ads.csv')

# Filter for Tehran residential sales (same as before)
df_text_source = df_original_reload[
    (df_original_reload['cat2_slug'] == 'residential-sell') & 
    (df_original_reload['city_slug'] == 'tehran')
].copy()

print(f"Reloaded data shape: {df_text_source.shape}")
print(f"Title column present: {'title' in df_text_source.columns}")
print(f"Description column present: {'description' in df_text_source.columns}")

# Create combined text
df_text_source['combined_text'] = df_text_source['title'].fillna('') + ' ' + df_text_source['description'].fillna('')

print(f"\nSample title: {df_text_source['title'].iloc[0][:100]}...")
print(f"Sample description: {df_text_source['description'].iloc[0][:100] if pd.notna(df_text_source['description'].iloc[0]) else 'None'}...")
```

```python
# ============================================
# Extract NLP Features from Full Dataset
# ============================================

print("Extracting NLP features from 94,541 listings... (this will take ~2-3 minutes)")
print("Processing...")

# Apply the extraction function
nlp_features_full = df_text_source['combined_text'].progress_apply(extract_nlp_features) if hasattr(pd.Series, 'progress_apply') else df_text_source['combined_text'].apply(extract_nlp_features)

print(f"\n✅ Extracted {nlp_features_full.shape[1]} features from {nlp_features_full.shape[0]} listings")

# Show feature prevalence for the full dataset
print("\n📊 Feature Prevalence (Full Dataset - 94,541 listings):")
print("=" * 65)
feature_counts_full = nlp_features_full.sum().sort_values(ascending=False)
for feature, count in feature_counts_full.items():
    pct = count / len(nlp_features_full) * 100
    bar = "█" * int(pct / 2)
    print(f"{feature:25} {count:6,} ({pct:5.1f}%) {bar}")
```

```python
# ============================================
# Merge NLP Features with df_tehran_smart
# ============================================

# Add NLP features to the text source dataframe
df_text_source_with_nlp = pd.concat([df_text_source.reset_index(drop=True), nlp_features_full.reset_index(drop=True)], axis=1)

print(f"Text source with NLP shape: {df_text_source_with_nlp.shape}")

# Now we need to match rows between df_tehran_smart and df_text_source_with_nlp
# We'll use a combination of key columns to create a unique identifier

# First, let's check what columns we can use as keys
key_cols = ['price_value', 'building_size', 'neighborhood_slug', 'construction_year', 'rooms_count']
print(f"\nChecking key columns in both dataframes...")

for col in key_cols:
    in_smart = col in df_tehran_smart.columns
    in_text = col in df_text_source_with_nlp.columns
    print(f"  {col}: smart={in_smart}, text={in_text}")
```

```python
# ============================================
# Create Matching Keys and Merge
# ============================================

# Create a composite key for matching
# Using multiple columns to ensure unique identification

nlp_feature_cols = [
    'is_north_facing', 'is_south_facing', 'is_east_facing', 'is_west_facing',
    'is_two_sided', 'is_well_lit', 'has_view', 'is_newly_built', 'is_renovated',
    'is_luxury', 'is_fully_equipped', 'is_never_used', 'is_dead_end',
    'has_good_layout', 'has_single_deed', 'has_roof_garden', 'has_lobby',
    'has_caretaker', 'has_video_intercom', 'has_built_in_closet', 'has_remote_door'
]

# Select only NLP features + key columns from text source
merge_cols = ['price_value', 'building_size', 'neighborhood_slug', 'floor', 'rooms_count'] + nlp_feature_cols
df_nlp_to_merge = df_text_source_with_nlp[merge_cols].copy()

# Remove duplicates based on key columns (keep first occurrence)
df_nlp_to_merge = df_nlp_to_merge.drop_duplicates(subset=['price_value', 'building_size', 'neighborhood_slug', 'floor', 'rooms_count'], keep='first')

print(f"NLP features to merge: {len(df_nlp_to_merge)} unique rows")

# Merge with df_tehran_smart
df_with_nlp = df_tehran_smart.merge(
    df_nlp_to_merge,
    on=['price_value', 'building_size', 'neighborhood_slug', 'floor', 'rooms_count'],
    how='left'
)

print(f"\n✅ Merged dataset shape: {df_with_nlp.shape}")
print(f"   Original df_tehran_smart: {df_tehran_smart.shape}")

# Check how many rows got NLP features
nlp_coverage = df_with_nlp[nlp_feature_cols].notna().all(axis=1).sum()
print(f"\n📊 NLP Feature Coverage: {nlp_coverage:,} / {len(df_with_nlp):,} ({nlp_coverage/len(df_with_nlp)*100:.1f}%)")
```

```python
# ============================================
# Fill Missing NLP Features and Finalize
# ============================================

# For rows that didn't match, fill NLP features with False (conservative approach)
for col in nlp_feature_cols:
    df_with_nlp[col] = df_with_nlp[col].fillna(False).astype(bool)

# Verify no missing values in NLP columns
missing_nlp = df_with_nlp[nlp_feature_cols].isna().sum().sum()
print(f"Missing values in NLP features: {missing_nlp}")

# Show updated feature prevalence
print("\n📊 NLP Feature Prevalence in Final Dataset:")
print("=" * 65)
for col in nlp_feature_cols:
    count = df_with_nlp[col].sum()
    pct = count / len(df_with_nlp) * 100
    bar = "█" * int(pct / 2)
    print(f"{col:25} {count:6,} ({pct:5.1f}%) {bar}")

# Update the main dataset
df_tehran_final = df_with_nlp.copy()

print(f"\n✅ Final dataset with NLP features: {df_tehran_final.shape}")
print(f"   Total columns: {len(df_tehran_final.columns)}")
print(f"   - Original features: 29")
print(f"   - New NLP features: 21")
```
